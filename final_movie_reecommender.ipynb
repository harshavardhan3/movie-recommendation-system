```python
#dont run this code on jupyter notebook as i observed theres a huge amount of ram usage sometimes Please be advised!
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import time # for calculating execution times
from sklearn import preprocessing #for min max normalisation

#importing required libraries
timea = time.time()
```

```python
min_max_scaler = preprocessing.MinMaxScaler()
movies = pd.read_csv('../input/the-movies-dataset/movies_metadata.csv')
keywords = pd.read_csv('../input/the-movies-dataset/keywords.csv')
credits = pd.read_csv('../input/the-movies-dataset/credits.csv')
#importing data from read_csv
```

```python
#function for removing unusual ids of movies
def to_int(x):
    try:
        return int(x)
    except:
        return np.nan
data = [movies['id'],movies['title'],credits['cast'],credits['crew'],keywords['keywords'],movies['genres'],movies['production_companies'],movies['original_language'],movies['overview'],movies['belongs_to_collection'],movies['popularity'],movies['vote_average'],movies['vote_count']]
data = pd.DataFrame(data).T
#Making pandas dataframe from the required data
data['id'] = data['id'].apply(to_int)
data = data[data['id'].notnull()]
#removing the unidentified ids
```

```python
def str_to_data(x):
    try:
        return eval(x,{"__builtins__":None})
    except:
        return []
features = ['cast','crew','keywords','genres','production_companies','belongs_to_collection']
for feature in features:
    data[feature]=data[feature].apply(str_to_data)
#making the string values to python datatypes in loop using powerful eval() tool inbuit in python for string parsing
#for more reference please visit
#https://www.journaldev.com/22504/python-eval-function
```

```python
features = ['popularity','vote_average','vote_count']
data[features] =pd.DataFrame(min_max_scaler.fit_transform(data[features].values))
#normalising the three columns in pandas which has float data of popularity, tmdb vote_count, tmdb vote_average
```

```python
#function to search for director and return the name
def get_director(total_crew):    
    for crew in total_crew:
        if crew['job'].lower() == 'director':
            return crew['name']
    return np.nan
data['director'] = data['crew'].apply(get_director)
#execution of above function
```

```python
#function to get top 3 cast names as list
def get_cast(total):
    if isinstance(total,list):
        names = [ele['name'] for ele in total]
        if len(names)>3:
            names = names[:3]
        return names
    return []
data['cast'] = data['cast'].apply(get_cast)
#execution of above function
```

```python
#function to get all the keywords/genres/production_companies as list
def get_list(total):
    if isinstance(total,list):
        names = [ele['name'] for ele in total]
        return names
    return []
data['keywords'] = data['keywords'].apply(get_list)
data['genres'] = data['genres'].apply(get_list)
data['production_companies'] = data['production_companies'].apply(get_list)
#execution of above function
```

```python
#function to get the collection as it is in dictionary data type
def get_collection(collection):
    if isinstance(collection,dict):
        names = [collection['name']]
        return names
    return []
data['belongs_to_collection'] = data['belongs_to_collection'].apply(get_collection)
#execution of above function
```

```python
#stripping by spaces of titles for searching and making it as single word/string
def strip(x):
    if isinstance(x,list):
        return [str.lower(i.replace(" ","")) for i in x]
    else:
        if isinstance(x,str):
            return str.lower(x.replace(" ",""))
        else:
            return ""
features = ['cast','director','keywords','genres','production_companies','belongs_to_collection']
for feature in features:
    data[feature]= data[feature].apply(strip)
#apply the function to all other dataframe lists/strings
data['title_dup'] = data['title'].apply(strip)
#creating duplicaate titles which has stripped original titles
```

```python
def metadata(x):
    return " ".join(x['keywords'])+" "+" ".join(x['genres'])+" "+" ".join(x['cast'])+" "+x['director']+" "+" ".join(x['production_companies'])+" "+" ".join(x['belongs_to_collection'])
data['metadata']=data.apply(metadata,axis = 1)
#creating metadata sentence which has all the details of the movie
```

```python
data['index'] = [i for i in range(len(data))]
#creating a index for data frame because it doesn't push if deleted any row
```

```python
indexing = pd.Series(data.index, index = data['index'])
indices = pd.Series(data.index, index = data['title_dup'])
#creating reverse indexes for getting the index for dataframe from data['index'] as well as from duplicate titles
```

```python
from sklearn.feature_extraction.text import CountVectorizer 

count = CountVectorizer(stop_words = 'english')
count_matrix = count.fit_transform(data['metadata'])
#making a count vectorizer matrix 
```

```python
count_matrix
# A sparse matrix 
```

```python
# recommender function
def metadata_recommender(title,count_matrix = count_matrix ,df = data, indices = indices,indexing = indexing,content_weight=1,user_weight=0.01,popularity_weight=1,vote_average_weight = 1, vote_count_weight = 1):
    title = str.lower(title.replace(" ","")) #strip the title to check it in duplicate titles
    idxs = indices[title] #get the indexes of title here if one movie exists it will give numpy.int64 if there are more movies it will give pandas.series of numpy.int64 a pandas dataframe
    movie_indices=[]
    movies_list = []
    movies_dict = []
    if isinstance(idxs,np.int64):
        idxs = [idxs] #if the idxs has only one np.int64 then it cant be looped in below so we make a list of that one value if its a list it can be iterated
    for idx in idxs: #for each idx in idxs
        features = ['popularity','vote_average','vote_count'] #non string features which has floats
        x = np.nan_to_num(data[features].values[data['index'][idx]]) #non string features of the title we want
        w = np.asarray([popularity_weight , vote_average_weight, vote_count_weight])
        y = np.nan_to_num(data[features].values) #total features for matching
        x = x*w #multiplying weights to each value of movie.
        math_scores = np.dot(x,y.T) # dot product for getting the scores of non string features
        math_scores = (min_max_scaler.fit_transform(math_scores.reshape(-1,1))).reshape(math_scores.shape[0]) #normalising it 
        sim_scores = (count_matrix[data['index'][idx]]*count_matrix.T).toarray()[0] #dot product of string features count array of the title we want
        # with the total features to get the cosine similarity
    
        sim_scores = (min_max_scaler.fit_transform(sim_scores.reshape(-1,1))).reshape(sim_scores.shape[0]) #normalization
        sim_scores = list(enumerate(sim_scores)) #enumerating and making it as tuple list
        sim_scores = sorted(sim_scores, key = lambda x:x[1],reverse = True) #sorting in descending order
        #print(sim_scores)
        #sim_score = sim_scores[1:21]
        sim_scores = sim_scores[1:51] #taking top 50 movies which has similar metadata features
    
        for i in range(len(sim_scores)):
            sim_scores[i] = (sim_scores[i][0],(content_weight*sim_scores[i][1]+user_weight*math_scores[sim_scores[i][0]])) #calculating weighted sum of metadat features and user rating features       
        sim_scores = sorted(sim_scores, key = lambda x:x[1],reverse = True) # again sorting them in order for getting combined effect
        sim_scores = sim_scores[0:20]#taking top 20 of combined list
        #content_indices = [i[0] for i in sim_score]
        #print(content_indices)
        #print(movie_indices)
        movie_indices.append([indexing[i[0]] for i in sim_scores]) #getting the movie indices
    for movie_index in movie_indices:
        movies_list.append(data['title'].iloc[movie_index])
        i=0
    for idx in idxs:
        movies_dict.append({"Overview" : data['overview'].iloc[data['index'][idx]],"Movies List": movies_list[i]})
        i=i+1
    return movies_dict #data['title'].iloc[movie_index for movie_index in movie_indices],timeb-timea #returning those movies
timeb = time.time()
```
